```python
from datautil.common import path_to_save_base
import pandas as pd
import numpy as np
import re

def roster_file_loader(path):
    fln = open(path,  encoding="utf-8").readlines()
    header = [l for l in fln if l.startswith('//id')][0]
    header = [tk.strip() for tk in header[2:].split(',')] 
    
    def auto_cast(t):
        try:
            return int(t)
        except:
            return t
    
    contents = [dict(zip(header,[auto_cast(t) for t in l.split(',')][:-1])) for l in fln if l.startswith('//') == False] # remove last column (eol)
    print (contents[0])
    
    df = pd.DataFrame(contents)[header] # to be sorted as original just in case
    return df.set_index('id')
```

```python
roster = roster_file_loader(path_to_save_base("New Game 2")+ r"/import_export/" + "fbl_rosters" + ".txt")
```
